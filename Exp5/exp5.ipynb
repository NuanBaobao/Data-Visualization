```python
# 2.导入相关库
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 3.定义树节点
treeNodes={}
class Node:
        def __init__(self,par_name=None,y=0,left_bro=None,name=None):
            self.width=0 # 树节点大小
            self.parent=par_name
            self.x=None
            self.y=y
            self.children=[]
            self.left_bro=left_bro # 树结点的左右兄弟
            # 树节点的最左右节点坐标值
            self.x_left={1:500,2:500,3:500,4:500} 
            self.x_right={1:-500,2:-500,3:-500,4:-500}
            self.offset=0 # 合并时移动步长
            self.name=name
        # 添加子节点
        def add_child(self,ch):
            self.children.append(ch)
        # 寻找左兄弟
        def get_lbrother(self):
            if self.left_bro == None:
                return None
            return treeNodes[self.left_bro]
        # 寻找父节点
        def get_parent(self):
            return treeNodes[self.parent]
```

```python
# 4.定义添加树节点函数
def appendNode(data,par_name=None,y=0,left_bro=None):
    name=data['name']
    # 将当前节点加入树节点
    if name not in treeNodes.keys():
        treeNodes[name]=Node(par_name,y,left_bro,name)
    else:
        return
    if 'children' in data.keys():
        children=data['children']
    else:
        return
    
    if len(children)!=0:
        left=None
        for t in children:
            # 遍历子节点
            treeNodes[name].add_child(t['name'])
            appendNode(t,name,treeNodes[name].y+1,left_bro=left)
            left=t['name']
            
appendNode(data)
```

```python
# 5.初始化
def initial(node):
    # 获取当前节点属性
    node.x=0
    y=node.y
    if len(node.children)==0:
        if node.y!=0:
            node.width=1/node.y # 节点宽度
            node.x_left[y]=0 # 初始化左节点坐标
            node.x_right[y]=node.width # 初始化右节点坐标
            node.x=node.width/2 # x坐标
    else:
        if node.y!=0:
            node.width=1/node.y
        # 遍历子节点，获取当前层节点合并需移动距离
        for k in range(len(node.children)):
            children=treeNodes[node.children[k]]
            initial(children)
            # 叶子节点
            if k==0:
                for i in range(y+1,5):
                    node.x_left[i]=children.x_left[i]
            for i in range(y+1,5):
                node.x_left[i]=min(node.x_left[i],children.x_left[i])
                node.x_right[i]=max(node.x_right[i],children.x_right[i])
        # 获取最左右节点坐标
        node.x_left[y]=(treeNodes[node.children[0]].x+treeNodes[node.children[-1]].x+treeNodes[node.children[-1]].offset)/2
        node.x_right[y]=node.x_left[y]+node.width
        node.x=node.x_left[y]
    left_bro=node.get_lbrother() # 获取左兄弟节点
    offset=0
    if left_bro!=None:
        offset=0
        for i in range(y,5):
            offset=max(offset,left_bro.x_right[i]-node.x_left[i]) # 更新移动距离

    for i in range(y,5):
        node.x_left[i]+=offset
        node.x_right[i]+=offset
    node.offset=offset
    
```

```python
# 6.移动节点
def update(node,disp):
    node.x+=node.offset+disp
    for i in node.children:
        # 递归子节点
        update(treeNodes[i], disp+node.offset)
    node.offset=0
    
```

```python
# 7.获取树的直径
max_x=0
for i in range(1,5):
    max_x=max(max_x,treeNodes['root'].x_right[i])
    print(max_x)
```

```python
# 8.绘制树节点
def draw_node(node,maxx,r):
    x,y=node.x,node.y
    angle = 2 * np.pi * x / (maxx) # 角度映射
    y, x = y * np.sin(angle)*r, y * np.cos(angle)*r
    # rotation, alignment = get_label_rotation(angle, np.pi/2)

    plt.scatter(x, y, facecolor='red',marker="o")
    plt.text(x, y, node.name, fontsize=10,)
    for child in node.children:
        draw_node(treeNodes[child], maxx,r)

```

```python
# 9.绘制连线
def draw_line(node, maxx,r):
    x,y=node.x,node.y
    angle = 2 * np.pi * node.x / (maxx)
    y, x = y * np.sin(angle)*r, y * np.cos(angle)*r
    for i in node.children:
        child=treeNodes[i]
        childx = child.x
        childy = child.y
        angle = 2 * np.pi * childx / (maxx)
        childy, childx = childy * np.sin(angle)*r, childy * np.cos(angle)*r
        plt.plot([x, childx], [y, childy]
        , linestyle='dashed', linewidth=1, color="green")
        draw_line(child, maxx,r)
```

```python
plt.figure(figsize=(20, 20))
draw_node(treeNodes['root'],max_x,1)
draw_line(treeNodes['root'],max_x,1)
plt.savefig('output.png')
plt.show()
```
